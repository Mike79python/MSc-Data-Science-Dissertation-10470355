```python
import threading
import requests
from bs4 import BeautifulSoup
import sqlite3

base_url = "https://bleacherreport.com/articles/"
ARTICLE_NUM_START = 360127
ARTICLE_NUM_END = 10056165
NUM_THREADS = 12
ARTICLES_PER_THREAD = (ARTICLE_NUM_END - ARTICLE_NUM_START) // NUM_THREADS


def get_article_data(soup: BeautifulSoup):
    title = soup.title.get_text()
    title = title[0 : title.find("|")]

    date = soup.find(class_="date")
    date = date.get_text(strip=True)

    p_tags = soup.find_all("p")
    article_text = ""
    for p in p_tags:
        article_text += p.text

    return title, date, article_text


def scrape_all(num_threads):
    conn = sqlite3.connect("results.db")
    cursor = conn.cursor()

    cursor.execute(
        """
        CREATE TABLE IF NOT EXISTS articles (
            id INTEGER PRIMARY KEY,
            title TEXT,
            breadcrumb TEXT,
            date TEXT,
            text TEXT
        )
    """
    )

    threads: list[threading.Thread] = []

    for i in range(num_threads):
        t = threading.Thread(target=scrape_worker, args=(i,))
        threads.append(t)
        t.start()

    for i in range(num_threads):
        threads[i].join()


def scrape_worker(id):
    start = ARTICLE_NUM_START + (id * ARTICLES_PER_THREAD)
    end = (
        ARTICLE_NUM_END if (id == NUM_THREADS - 1) else start + ARTICLES_PER_THREAD - 1
    )

    conn = sqlite3.connect("results.db")
    cursor = conn.cursor()
    for article_num in range(start, end):
        if article_num % 100 == 0:
            print(f"Thread {id} at article {article_num}...")
        cursor.execute("SELECT id FROM articles WHERE id = ?", (article_num,))
        existing_article = cursor.fetchone()
        if existing_article:
            print("Already exists")
            continue

        # Article does not exist, proceed with scraping
        resp = requests.get(base_url + str(article_num))
        soup = BeautifulSoup(resp.text, "html.parser")

        element = soup.find(class_="organism nr-breadcrumb")
        if element:
            breadcrumbs = element.get_text(strip=True)
            if breadcrumbs.__contains__("Formula 1"):
                title, date, article_text = get_article_data(soup=soup)
                cursor.execute(
                    "INSERT INTO articles VALUES (?, ?, ?, ?, ?)",
                    (article_num, title, breadcrumbs, date, article_text),
                )
                print(f"Added article {article_num} with title {title}")

        conn.commit()
    conn.close()


scrape_all(NUM_THREADS)
```
