# **Machine Learning Algorithms implementation and optimization**

**Import necessary libraries**


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, roc_curve, auc, roc_auc_score
import warnings
from sklearn.exceptions import ConvergenceWarning
import random
from sklearn.model_selection import cross_val_score
warnings.filterwarnings("ignore", category=ConvergenceWarning)
import matplotlib.pyplot as plt
```

---
**Import data and split into training and validating sets**

```python
'''
Data is loaded from the pre-processing section. 
Here Rank is limited to 20 places, since 2017, as the teams are now 10. 
In some seasons there were more, but we are interested in the first 20 only.
'''

data = pd.read_csv('df1.csv')
pd.set_option('display.max_columns', None)
df = data.copy()
df = df[df['Rank'] <= 20]
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
'''
Data is split into training and validation sets. Training is data from the seasons from 2010 to 2021, 
and validation is data from the season 2022. Rank is the target variable.

'''


test = df[df['season'] == 2022].drop('season', axis = 1)
y_test = test['Rank']
x_test = test.drop('Rank', axis = 1)

train = df[df['season'] < 2022].drop('season', axis = 1)
y_train = train['Rank']
x_train = train.drop('Rank', axis = 1)
```

---
**Machine Learning Algorithms Implementation**

```python
RANDOM_SEED = 42

classifiers = {
    'ExtraTreesClassifier': ExtraTreesClassifier(random_state=RANDOM_SEED),
    'MLPClassifier' : MLPClassifier(random_state = RANDOM_SEED),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=RANDOM_SEED),
    "LogisticRegression": LogisticRegression(random_state=RANDOM_SEED),
    "SVM": SVC(probability=True, random_state=RANDOM_SEED),
    'GaussianNB': GaussianNB(),
    "KNeighborsClassifier": KNeighborsClassifier(5),
    "RandomForestClassifier": RandomForestClassifier(random_state=RANDOM_SEED),
    "AdaBoostClassifier": AdaBoostClassifier(random_state=RANDOM_SEED),
    "XGBoostClassifier": XGBClassifier(random_state=RANDOM_SEED)
}
```

```python
'''
This functions consist of the scoring functions, 
due to the complexity of the domain of predicting all
the ranks of each race in a season, the scorer function, accepts missclassification 
of up to 2 places above or below. Alternative_scorer, is the Upper Bound Accuracy scorer, 
that accepts correct classification, if the predicted class is lower than the actual.

'''

def scorer(row, threshold):
    if threshold-2 <= row <= threshold+2:
        return True
    else:
        return False

def alternative_scorer(row, threshold):
    if threshold-1 <= row <= threshold+20:
        return True
    else:
        return False

def np_scorer(row, threshold):
    return np.logical_and(np.array(threshold) - 2 <= row, row <= np.array(threshold) + 2)
```

```python
'''
This function consists of the scores of each model. here all the models are tested
in overall accuracy, UBA accuracy, AUC-ROC and winners predictions.
'''


def class_score(x_train, y_train, x_test,y_test):
    performance = {}
    lb = LabelBinarizer()
    for name, classifier in classifiers.items():
        print(name)
        if name == "XGBoostClassifier":
            y_train = y_train-1
            y_test = y_test-1
        else:
            None

        classifier.fit(x_train, y_train)
        preds = classifier.predict(x_test)

        pred_df = pd.DataFrame(preds,np.array(y_test)).reset_index().rename(columns = {'index' : 'Actual', 0 : 'Prediction'})
        pred_df['Score'] = pred_df.apply(lambda row : scorer(row[1],row[0]), axis = 1)
        pred_df['Worst Case Scorer']= pred_df.apply(lambda row : alternative_scorer(row[1],row[0]), axis = 1)

        true =sum(pred_df['Score'] == True)
        false = sum(pred_df['Score'] == False)

        true_worst = sum(pred_df['Worst Case Scorer'] == True)
        false_worst = sum(pred_df['Worst Case Scorer'] == False)

        accuracy = true/(true+false)

        alt_accuracy = true_worst/(true_worst+false_worst)

        first_perc = sum(pred_df[pred_df['Actual']==1]['Prediction'] == 1)
        pred_df['corrected'] = pred_df.apply(lambda row : row[0] if row[2] is True else row[1], axis = 1)
        y_true = np.array(pred_df['Actual'])
        y_pred = np.array(pred_df['corrected'])
        y_true_binary = lb.fit_transform(y_true)
        y_pred_corrected_binary = lb.transform(y_pred)
        y_pred_corrected_binary
        auc_roc = roc_auc_score(y_true_binary, y_pred_corrected_binary)
        performance[name] = [accuracy,alt_accuracy, auc_roc, first_perc/22]


    return pd.DataFrame(performance, index = ['Accuracy', 'Upper Bound Accuracy' ,'AUCROC', 'First Place Accuracy']).T

class_score(x_train, y_train, x_test,y_test)
```

---
**Hyperparameter Optimisation**


```python
'''
This scorer is used for the hyper parameter optimization, using the custom overall score from before.
'''
def single_class_score(x_train, y_train, x_test,y_test, classifier):

        classifier.fit(x_train, y_train)
        preds = classifier.predict(x_test)

        pred_df = pd.DataFrame(preds,np.array(y_test)).reset_index().rename(columns = {'index' : 'Actual', 0 : 'Prediction'})
        pred_df['Score'] = pred_df.apply(lambda row : scorer(row[1],row[0]), axis = 1)
        true = sum(pred_df['Score'] == True)
        false = sum(pred_df['Score'] == False)
        accuracy = true/(true+false)

        return accuracy
```

```python
'''
Customised Random Search for hyperparmeter optimization.
'''

def Randomised_Search(y_test, y_train, classifier, param_grid, n=50):

    best_precision = 0  

    for _ in range(n):
        hyperpar_used = {i: random.choice(j) for i, j in param_grid.items()}
        if classifier == 'ExtraTreesClassifier':
            classif = ExtraTreesClassifier(**hyperpar_used, random_state=42)
        elif classifier == 'GradientBoostingClassifier':
            classif = GradientBoostingClassifier(**hyperpar_used, random_state=42)
        elif classifier == "RandomForestClassifier":
            classif = RandomForestClassifier(**hyperpar_used, random_state=42)
        elif classifier == 'AdaBoostClassifier':
            classif = AdaBoostClassifier(**hyperpar_used, random_state=42)
        elif classifier == 'XGBoostClassifier':
            classif = XGBClassifier(**hyperpar_used, random_state=42) 

        precision = single_class_score(x_train, y_train, x_test, y_test, classif)
        if precision > best_precision:
            best_precision = precision
            print('Best Accuracy so far:', precision)
            best_hyp = hyperpar_used
    print('Best Accuracy :', best_precision)
    return best_hyp
```
