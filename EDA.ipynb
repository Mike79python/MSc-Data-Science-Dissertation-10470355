```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from matplotlib.ticker import MaxNLocator

merged = pd.read_csv('mdf7.csv')


pd.set_option('display.max_columns', None)
```

Most succesive driver in the period 2010 and 2021 

```python
drivers_rank1 = merged[merged['Rank'] == 1]['driver_id'].value_counts()
drivers_grid1 = merged[merged['grid'] == 1]['driver_id'].value_counts()

top_10_rank1 = drivers_rank1.nlargest(10)
top_10_grid1 = drivers_grid1.nlargest(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10_rank1.index, top_10_rank1.values)
plt.xlabel('Driver')
plt.ylabel('Wins')
plt.title('Most Wins from 2010 to 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(top_10_grid1.index, top_10_grid1.values)
plt.xlabel('Driver')
plt.ylabel('Pole positions')
plt.title('Most pole positions from 2020 to 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()





```

```python
# Assuming merged is your DataFrame after previous operations

circuits_df = merged[['season', 'circuit_id']]
circuits_df.drop_duplicates(inplace =True)
circuits_df.reset_index(inplace = True, drop = True)
circuit_counts = circuits_df['circuit_id'].value_counts().reset_index()
circuit_counts.columns = ['circuit_id', 'count']
# Filter the original DataFrame for rows where grid=1 and Rank=1 and group by circuit_id
win_from_pole = merged[(merged['grid'] == 1) & (merged['Rank'] == 1)]
win_from_pole_counts = win_from_pole.groupby('circuit_id').size().reset_index(name='win_from_pole_count')

# Merge with the circuit_counts dataframe
result = pd.merge(circuit_counts, win_from_pole_counts, on='circuit_id', how='left')

# Fill NaN values with 0 (there might be circuits with no wins from pole)
result['win_from_pole_count'].fillna(0, inplace=True)

# Calculate the percentage
result['percentage'] = (result['win_from_pole_count'] / result['count']) * 100
# Sorting the data for better visualization
result = result.sort_values(by='percentage', ascending=False)

plt.figure(figsize=(15, 10))
plt.bar(result['circuit_id'], result['percentage'])
plt.ylabel('Percentage (%)')
plt.xlabel('Circuit ID')
plt.title('Percentage of Races Won from Pole for Each Circuit')
plt.xticks(rotation=90)  # Rotate circuit names for better visibility
plt.show()
```

```python
data_first3 = merged.query('Rank in [1, 2, 3 ]').copy()

# Single plot
fig, ax = plt.subplots(figsize=(18, 10))

# Only the KDE plot
sns.kdeplot(data=data_first3, x='salary', hue='Rank', ax=ax, palette=['red', 'green', 'blue'], fill=True)
ax.set_xlim([0, 50000000])

plt.tight_layout()  # To ensure plots don't overlap
plt.show()
```

```python
# Filter for home races
home_races = merged[merged['driver_home_race'] == 1]

# Group by circuit ID, and calculate the average rank
avg_ranks = home_races.groupby('circuit_id')['Rank'].mean().reset_index()

# Plot the data
plt.figure(figsize=(15, 6))
sns.barplot(x='circuit_id', y='Rank', data=avg_ranks, palette='viridis')
plt.title('Average Rank at Home Races for Each Circuit')
plt.xlabel('Circuit ID')
plt.ylabel('Average Rank')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
